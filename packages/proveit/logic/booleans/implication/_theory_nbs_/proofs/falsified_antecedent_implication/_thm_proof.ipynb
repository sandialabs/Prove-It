```python
import proveit
theory = proveit.Theory() # the theorem's theory
from proveit import defaults, display_provers # useful imports
defaults.assumptions = [] # often useful to set
defaults.auto_simplify = False # can be convenient to enable
from proveit import A, B
from proveit.logic import Implies, Not, FALSE
```

```python
%proving falsified_antecedent_implication
```

```python
A_impl_F = Not(A).unfold(assumptions=[Not(A)])
```

```python
A_impl_F.derive_consequent(assumptions=[A])
```

```python
F_impl_B = Implies(FALSE, B).prove()
```

```python
B_via_contradiction = F_impl_B.derive_consequent(assumptions=[Not(A), A])
```

```python
B_via_contradiction.as_implication(A)
```
