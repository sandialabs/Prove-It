These are all different styles for the same expression.  The condition style may be 'compact' or 'expanded' and the parameter variable is also inconsequential (considered to be a special kind of style option):

```python
forall_Py_suchthat_Qy == forall_Py_suchthat_Qy__restyled == forall__Qx_impl_Px == forall__Qx_impl_Px__restyled
```

## Instantiation

Universally quantified statements are particularly valuable because they may be instantiated to any valid expressions satisfying conditions.  The `instantiate` method has a behavior that depends on the style.

```python
%%time
from proveit import Function
from proveit.logic import InSet
from proveit.numbers import Add, one
zp1 = Add(z, one)
```

With condition='compact' style, the `instantiate` will automatically derive the consequent of the internal implication via modus ponens.

```python
P_zp1 = forall_Py_suchthat_Qy.instantiate({y:zp1}, assumptions=[forall_Py_suchthat_Qy, Function(Q, zp1)])
```

```python
P_zp1.proof()
```

This is with the condition='expanded' style: 

```python
%%time
forall__Qx_impl_Px.instantiate({x:zp1}, assumptions=[forall__Qx_impl_Px])
```
