```python
import proveit
theory = proveit.Theory() # the theorem's theory
from proveit import defaults, display_provers # useful imports
defaults.assumptions = [] # often useful to set
defaults.auto_simplify = False # can be convenient to enable
from proveit import A
from proveit.logic import Not
```

```python
%proving forall_by_excluded_middle
```

```python
defaults.assumptions = forall_by_excluded_middle.instance_expr.antecedent.operands
```

```python
from proveit.logic.equality import equals_reflexivity
```

```python
from proveit import PofA, x
```

```python
equals_reflexivity.instantiate({x:PofA}, preserve_all=True)
```

```python
PofA_by_A = defaults.assumptions[0].instantiate(append_assumptions=[A])
```

```python
PofA_by_notA = defaults.assumptions[1].instantiate(append_assumptions=[Not(A)])
```

```python
PofA = PofA_by_A.as_implication(A).derive_consequent_via_excluded_middle()
```

Automatic generalization to prove the theorem doesn't work since the theorem uses `B` instead of `A`.  But once we generalize $\forall_A P(A)$, $\forall_B P(B)$ is known automatically (they are equal expressions in Prove-It).

```python
PofA.generalize(A)
```
