```python
import proveit
theory = proveit.Theory() # the theorem's theory
from proveit import defaults, display_provers # useful imports
defaults.assumptions = [] # often useful to set
defaults.auto_simplify = False # can be convenient to enable
from proveit import A
from proveit.logic import Equals, TRUE
```

```python
%proving untrue_from_negation
```

```python
defaults.assumptions = untrue_from_negation.conditions + (Equals(A, TRUE),)
```

```python
A_impl_F = defaults.assumptions[0].unfold()
```

```python
contradiction = A_impl_F.derive_consequent()
```

```python
contradiction_implication = contradiction.as_implication(Equals(A, TRUE))
```

```python
contradiction_implication.derive_via_contradiction()
```
