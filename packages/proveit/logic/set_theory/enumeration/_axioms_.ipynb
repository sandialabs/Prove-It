** Membership in an enumeration is the same as being equal to any of the contents of the enumeration:**

```python
enumSetDef = Forall(l, Forall((x, yIter1l), 
                              Equals(InSet(x, Set(yIter1l)), 
                                     x_equals_any_y)),
                   domain=Naturals)
```
