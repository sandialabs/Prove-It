### Tuple lengths and length reductions for 1-9

```python
def count_from_one(count):
    return [num(k) for k in range(1, count+1)]
count_to_1_range, count_to_2_range, count_to_3_range, count_to_4_range, count_to_5_range, count_to_6_range, count_to_7_range, count_to_8_range, count_to_9_range = \
    [Equals(count_from_one(count), [ExprRange(k, k, one, num(count))]) for count in range(1, 10)]
```

```python
cur_vars = vars()
def nvars(count):
    return [cur_vars[chr(ord('a')+k)] for k in range(count)]
tuple_len_1_typical_equiv, tuple_len_2_typical_equiv, tuple_len_3_typical_equiv, tuple_len_4_typical_equiv, tuple_len_5_typical_equiv, tuple_len_6_typical_equiv, tuple_len_7_typical_equiv, tuple_len_8_typical_equiv, tuple_len_9_typical_equiv = \
    [Forall(nvars(count), Equals(Len(nvars(count)), Len([ExprRange(k, k, one, num(count))]))) for count in range(1, 10)]
```
